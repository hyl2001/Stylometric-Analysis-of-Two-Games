```python
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
df_auc = pd.read_excel('model_eval_res\\auc.xlsx')
df_acc = pd.read_excel('model_eval_res\\acc.xlsx')
df_pre = pd.read_excel('model_eval_res\\pre.xlsx')
df_rec = pd.read_excel('model_eval_res\\rec.xlsx')
df_f1s = pd.read_excel('model_eval_res\\f1s.xlsx')

names = [
    "Nearest Neighbors",
    "Linear SVM",
    "RBF SVM",
    "Decision Tree",
    "Random Forest",
    "Neural Net",
    "AdaBoost",
    "Naive Bayes",
    "XGBoost"
]
```

```python
fig = plt.figure(figsize=(8, 8))

ele_w = .5

for idx, name in enumerate(names):
    ax = fig.add_subplot(3, 3, idx+1)

    ax.set_title(name)
    ax.boxplot(df_auc[name], labels=['AUC'], positions=[0], widths=ele_w)
    ax.boxplot(df_acc[name], labels=['Acc'], positions=[1], widths=ele_w)
    ax.boxplot(df_pre[name], labels=['Pre'], positions=[2], widths=ele_w)
    ax.boxplot(df_rec[name], labels=['Rec'], positions=[3], widths=ele_w)
    ax.boxplot(df_f1s[name], labels=['F1S'], positions=[4], widths=ele_w)
    ax.tick_params(axis="x",direction="in")

plt.tight_layout()
```
